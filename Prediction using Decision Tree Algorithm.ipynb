## Task 3 : Prediction Using Decision Tree Algorithm

In this task, we have to Create the Decision Tree classifier and visualize it graphically. The purpose is if we feed any new data to this classifier, it would be able to predict the right class accordingly

```python
# Importing all libraries that we are going to use in this notebook

import pandas as pd                 # for data manipulation
import numpy as np                  # for numerical computation
import matplotlib.pyplot as plt     # for plotting graphs
import seaborn as sns               # for plotting charts

# Importing the libraries from scikit-learn package

from sklearn import datasets                               # for iris dataset
from sklearn.tree import DecisionTreeClassifier            # for using decision tree algorithm 
```

```python
# Loading the iris dataset

iris=datasets.load_iris()
```

```python
# Taking a look at data

iris.data
```

```python
# Name of the features present in the dataset

iris.feature_names
```

```python
# Looking at the targets

iris.target
```

```python
# Name of the targets

iris.target_names
```

```python
# Making a pandas dataframe

iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)

# Taking a look at final dataframe

iris_df.head()
```

```python
# Storing all targets in seprate variable y 

y = iris.target
```

```python
# Checking the number of rows and columns in our dataset

iris_df.shape
```

```python
# Checking the numerical info of all numerical columns in our dataset

iris_df.describe()
```

```python
# Checking basic info about all columns in our dataset

iris_df.info()
```

```python
# Plotting a graph for all 4 features

iris_df.plot(kind='line')
```

## Creating & Training Decision Tree Model

```python
# Defining the decision tree algorithm

model=DecisionTreeClassifier()
```

```python
# Training our Decision Tree Model using Iris Dataset

model.fit(iris_df,y)
```

## Visualizing Our Decision Tree

```python
# Checking the maximum depth of our decision tree

model.tree_.max_depth
```

```python
from sklearn.tree import plot_tree, export_text
plt.figure(figsize=(80,50))
plot_tree(model, feature_names=iris_df.columns,filled=True)
```

```python
# Visualizing our decision tree in text format

tree_text = export_text(model,max_depth=5,feature_names=list(iris_df.columns))
print(tree_text[:5000])
```

## Feature Importance

 Based on the gini index computations, a decision tree assigns importance value to each feature. These values can be used to intercept the results given by a decision tree.

```python
# Lets check out the model importance for each features

model.feature_importances_
```

```python
imp_df = pd.DataFrame({
    'feature':iris_df.columns,
    'importance':model.feature_importances_
}).sort_values('importance',ascending=False)
```

```python
imp_df
```

```python
plt.title('Feature Importance')
sns.barplot(data=imp_df,x='importance',y='feature')
```

## Predicting using Decision Tree Classifier of New Data

Let's make up some new data on our own and pass it to our Decision Tree Classifier model

```python
# Predicting the class of iris based on any new data

test = model.predict([[1.2,0.9,2.1,1.5],[1.3,2.5,0.9,0.8],[2.1,3.2,0.2,0.5]])
test
```
